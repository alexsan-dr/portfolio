# Шаг 2. Предобработка данных и подготовка их к исследованию

```python
def clean_text(series):
    return (
        series.astype('string')
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )

for df in [orders, events, tenge]:
    obj_cols = df.select_dtypes(include=['object', 'string']).columns
    for col in obj_cols:
        df[col] = clean_text(df[col])

for col in ['currency_code', 'device_type_canonical', 'cinema_circuit', 'service_name']:
    if col in orders.columns:
        orders[col] = orders[col].str.lower()

tenge['cdx'] = tenge['cdx'].str.lower()
```

```python
orders['created_dt_msk'] = pd.to_datetime(orders['created_dt_msk'], errors='coerce')
orders['created_ts_msk'] = pd.to_datetime(orders['created_ts_msk'], errors='coerce')
tenge['data'] = pd.to_datetime(tenge['data'], errors='coerce')

display(orders[['created_dt_msk', 'created_ts_msk']].head())
display(tenge[['data']].head())
```

```python
def plot_revenue_by_currency(df, currency):
    data = df.loc[df['currency_code'] == currency, 'revenue'].dropna()
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    
    axes[0].hist(data, bins=50)
    axes[0].set_title(f'Гистограмма revenue ({currency})')
    axes[0].set_xlabel('Выручка')
    axes[0].set_ylabel('Частота')
    
    axes[1].boxplot(data, vert=False)
    axes[1].set_title(f'Диаграмма размаха revenue ({currency})')
    axes[1].set_xlabel('Выручка')
    
    plt.show()

for cur in orders_df['currency_code'].dropna().unique():
    plot_revenue_by_currency(orders, cur)
```

```python
tickets = orders.merge(
    events,
    on='event_id',
    how='left',
    validate='m:1',
    indicator='events_merge_flag'
)

tickets['events_merge_flag'].value_counts()
```

```python
tenge_rate = tenge[['data', 'nominal', 'curs']].rename(columns={'data': 'created_dt_msk'})

tickets = tickets.merge(
    tenge_rate,
    on='created_dt_msk',
    how='left',
    validate='m:1'
)

tickets['revenue_rub'] = np.where(
    tickets['currency_code'] == 'rub',
    tickets['revenue'],
    tickets['revenue'] * tickets['curs'] / tickets['nominal']
)

tickets['one_ticket_revenue_rub'] = tickets['revenue_rub'] / tickets['tickets_count']

tickets['month'] = tickets['created_dt_msk'].dt.month

season_map = {
    12: 'зима', 1: 'зима', 2: 'зима',
    3: 'весна', 4: 'весна', 5: 'весна',
    6: 'лето', 7: 'лето', 8: 'лето',
    9: 'осень', 10: 'осень', 11: 'осень'
}
tickets['season'] = tickets['month'].map(season_map)

tickets['revenue_rub'] = tickets['revenue_rub'].round(2)
tickets['one_ticket_revenue_rub'] = tickets['one_ticket_revenue_rub'].round(2)

display(
    tickets[['currency_code', 'revenue', 'revenue_rub', 'tickets_count',
                'one_ticket_revenue_rub', 'month', 'season']].head()
)
```

**Сравнение структуры заказов летом и осенью по сегментам**

```python
season_df = eda_df[eda_df['season'].isin(['лето', 'осень'])].copy()
print(season_df['season'].value_counts())
```

```python
def build_season_share_table(df, segment_col):
    temp = (
        df.dropna(subset=[segment_col])
        .groupby(['season', segment_col])
        .agg(orders_count=('order_id', 'nunique'))
        .reset_index()
    )
    temp['share'] = temp['orders_count'] / temp.groupby('season')['orders_count'].transform('sum')
    
    pivot = temp.pivot(index=segment_col, columns='season', values='share').fillna(0)
    pivot = pivot[['лето', 'осень']] if set(['лето', 'осень']).issubset(pivot.columns) else pivot
    pivot['diff_pp'] = (pivot['осень'] - pivot['лето']) * 100
    pivot = pivot.sort_values('осень', ascending=False)
    return pivot

def plot_season_shares(pivot_table, title, xlabel='Доля заказов'):
    plot_df = pivot_table[['лето', 'осень']] * 100
    
    ax = plot_df.plot(kind='bar', figsize=(12, 5))
    ax.set_title(title)
    ax.set_xlabel('')
    ax.set_ylabel(xlabel)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Сезон')
    plt.tight_layout()
    plt.show()
```

<h2> Комментарий ревьюера v1 <a class="tocSkip"> </h2>

```python
event_type_shares = build_season_share_table(season_df, 'event_type_main')
display((event_type_shares * 100).round(2))
plot_season_shares(event_type_shares, 'Доли заказов по типам мероприятий: лето vs осень')
```

```python
mobile_orders = orders_per_user.loc[
    orders_per_user['device_type_canonical'] == 'mobile', 'orders_per_user'
]
desktop_orders = orders_per_user.loc[
    orders_per_user['device_type_canonical'] == 'desktop', 'orders_per_user'
]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(mobile_orders, bins=30, alpha=0.7, label='mobile')
axes[0].hist(desktop_orders, bins=30, alpha=0.7, label='desktop')
axes[0].set_title('Распределение числа заказов на пользователя')
axes[0].set_xlabel('Количество заказов')
axes[0].set_ylabel('Частота')
axes[0].legend()

axes[1].boxplot([mobile_orders, desktop_orders], labels=['mobile', 'desktop'])
axes[1].set_title('Диаграмма размаха числа заказов на пользователя')
axes[1].set_ylabel('Количество заказов')

plt.tight_layout()
plt.show()
```

Распределение количества заказов на пользователя является дискретным и, как правило, асимметричным, с выбросами и длинным правым хвостом.  
Поэтому для сравнения двух независимых выборок используется непараметрический тест Манна—Уитни.

Проверяем гипотезы:

- **H0**: количество заказов на одного пользователя у mobile **не выше**, чем у desktop;
- **H1**: количество заказов на одного пользователя у mobile **выше**, чем у desktop.

Уровень значимости: α = 0.05.

<div class="alert alert-success">
<h2> Комментарий ревьюера v1 <a class="tocSkip"> </h2>

<b>Все супер!👍:</b>Гипотезы точно сформулрованы и тип теста грамотно описан

```python
alpha = 0.05

orders_test = stats.mannwhitneyu(
    mobile_orders,
    desktop_orders,
    alternative='greater'
)

print('Статистика теста:', orders_test.statistic)
print('p-value:', orders_test.pvalue)

if orders_test.pvalue < alpha:
    print('Отвергаем H0: есть основания считать, что у mobile число заказов на пользователя выше.')
else:
    print('Не отвергаем H0: статистически значимых оснований считать, что у mobile число заказов выше, нет.')
```

```python
mobile_days = days_per_user.loc[
    days_per_user['device_type_canonical'] == 'mobile', 'mean_days_since_prev'
]
desktop_days = days_per_user.loc[
    days_per_user['device_type_canonical'] == 'desktop', 'mean_days_since_prev'
]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(mobile_days, bins=30, alpha=0.7, label='mobile')
axes[0].hist(desktop_days, bins=30, alpha=0.7, label='desktop')
axes[0].set_title('Распределение среднего интервала между заказами')
axes[0].set_xlabel('Дней')
axes[0].set_ylabel('Частота')
axes[0].legend()

axes[1].boxplot([mobile_days, desktop_days], labels=['mobile', 'desktop'])
axes[1].set_title('Диаграмма размаха среднего интервала между заказами')
axes[1].set_ylabel('Дней')

plt.tight_layout()
plt.show()
```

Распределение среднего интервала между заказами также может быть скошенным и содержать выбросы, поэтому для сравнения двух независимых групп снова используем тест Манна—Уитни.

Проверяем гипотезы:

- **H0**: среднее время между заказами у mobile **не выше**, чем у desktop;
- **H1**: среднее время между заказами у mobile **выше**, чем у desktop.

Уровень значимости: α = 0.05.

```python
days_test = stats.mannwhitneyu(
    mobile_days,
    desktop_days,
    alternative='greater'
)

print('Статистика теста:', days_test.statistic)
print('p-value:', days_test.pvalue)

if days_test.pvalue < alpha:
    print('Отвергаем H0: есть основания считать, что у mobile интервал между заказами выше.')
else:
    print('Не отвергаем H0: статистически значимых оснований считать, что у mobile интервал между заказами выше, нет.')
```
